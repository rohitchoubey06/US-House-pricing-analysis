```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
df2.groupby(['DATE'])['CSUSHPISA'].sum().plot(kind='bar')
plt.ylabel('Price in US$')
plt.title('US Housing price Fluctuation over years')
plt.show()
```

#### We can clearly see that since 1987 the Housing prices increased till 2006 and after that it declined till 2011 but after that the graph again took a rise and it shows a great hike from 2020-2021 and it is looking like it will head in the same direction for year 2022 as well.

### Now let's see the continue flow of Prices over the years 

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=df2.DATE,y=df2.CSUSHPISA,data=df2)
plt.show()
```

### The above graph shows a significant increase in rate of Price fluctuation after 2020 which continues to the present time that is 2022 and it's looking to follow the same pattern in future.

### Now let's see the same thing with a barplot

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df2.DATE,y=df2.CSUSHPISA,data=df2)
plt.show()
```
